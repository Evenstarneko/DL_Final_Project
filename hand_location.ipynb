```python
import numpy
import matplotlib.pyplot as plt
import pandas
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Flatten
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras import regularizers
from tensorflow.keras import initializers

# create and fit the LSTM network
reg=Sequential()
reg.add(LSTM(128,return_sequences=True,input_shape=(x_train.shape[1],3)))
reg.add(Dropout(0.15))
reg.add(LSTM(256,return_sequences=True))
reg.add(Dropout(0.15))
reg.add(LSTM(512,return_sequences=True))
reg.add(Dropout(0.25))
reg.add(Flatten())
reg.add(Dense(2))

reg.compile(optimizer="adam",loss="mean_squared_error")
reg.fit(x_train,y_train,epochs=100,batch_size=256)
```

```python
predict = reg.predict(x_train)
predict = predict.reshape((3, -1, 2))
target = y_train.reshape((3, -1, 2))

plt.plot(target[1, 0:25, 0], label="target")
plt.plot(predict[1, 0:25, 0], label="predict")
plt.xlabel("Trials")
plt.ylabel("Relative position")
plt.title("Predicted hand coordinates from RNN vs. ground truth (Training set)")
plt.legend()
```

```python
predictf = predict.flatten()
targetf = target.flatten()
MSE = np.sum((predictf - targetf)**2)/targetf.shape[0]
print(MSE)
```
